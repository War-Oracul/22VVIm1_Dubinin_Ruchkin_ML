```python
import pandas as pd
import seaborn as sns
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

<h3>Загрузка датасета</h3>

```python
df=pd.read_csv('CO2_emission.csv')
values = [ 'Pickup truck: Small','Pickup truck: Standard','Pickup truck: Standard','Station wagon: Mid-size','Special purpose vehicle','Station wagon: Small','Subcompact','Two-seater','Van: Passenger']

#drop rows that contain any value in the list
df = df[df.Vehicle_Class.isin (values) == False ]
#df["Vehicle_Class"] = np.where(df["Vehicle_Class"] == "SUV: Small", 0, 1)
df = df.drop(columns=['Transmission'])
df
```

<h3>Вывод основных статистических характеристик данных</h3>

```python
print(df.describe())
```

<h3>Корреляции</h3>

```python
df.corr()
```

```python
# Проверка баланса выходной переменной
def balanse_view(df_new):
    target_count = df_new.Vehicle_Class.value_counts()
    print('Class 0:', target_count[0])
    print('Class 1:', target_count[1])
    print('Пропорция:', round(target_count[0] / target_count[1], 2), ': 1')

    sns.countplot(x='Vehicle_Class', data= df_new, palette="OrRd")
    plt.box(False)
    plt.xlabel('Vehicle_Class 1/0',fontsize=11)
    plt.ylabel('Количество наблюдений',fontsize=11)
    plt.show()
```

<h3>Проверка баланса выходной переменной</h3>

```python
balanse_view(df)
```

<h3>Входные и выходные переменные</h3>

```python
X=df.iloc[:,4:11]
y=df['Vehicle_Class']
X
```

<h3>Андерсэмплинг с использованием TomekLinks</h3>

```python
from imblearn.under_sampling import TomekLinks

tl = TomekLinks(sampling_strategy='majority')
X_res, y_res = tl.fit_resample(X, y)
print(X_res.shape)
print(X.shape)
df_new=pd.concat([X_res, y_res], axis=1)
df_new.shape
```

```python
balanse_view(df_new)
```

<h3>Оверсэмплиг SMOT</h3>

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X, y)
print(X_sm.shape)
print(X.shape)
df_new_sm=pd.concat([X_sm, y_sm], axis=1)
df_new_sm.shape

sns.countplot(df_new_sm.Vehicle_Class, palette="OrRd")
plt.box(False)
plt.xlabel('Vehicle_Class 1/0',fontsize=11)
plt.ylabel('Количество наблюдений',fontsize=11)
plt.show()
```

```python
balanse_view(df_new_sm)
```

<h3>Случайный оверсэмплинг</h3>

```python
from imblearn.over_sampling import RandomOverSampler
autopct = "%.2f"
sampling_strategy = 'all'
ros = RandomOverSampler(sampling_strategy=sampling_strategy)
X_s, y_s = ros.fit_resample(X, y)
ax = y_s.value_counts().plot.pie(autopct=autopct)
_ = ax.set_title("Over-sampling")
```
