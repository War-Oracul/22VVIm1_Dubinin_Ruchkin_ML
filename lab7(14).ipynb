# Лабораторная работа 7. Кодирование качественных признаков

```python
import pandas as pd
```

```python
df=pd.read_csv('CO2_emission.csv')
cols_drop = ['Model_Year','Make','Model', 'Transmission', 'Vehicle_Class']
df = df.drop(columns=cols_drop)
df.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
X_train = df.iloc[:,:4]
X_train
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
print('Мат ожидание', scaler.mean_)
print('Дисперсия', scaler.var_)
print('Преобразованный набор')
arr = scaler.transform(X_train)
DfTr = pd.DataFrame(arr, columns=['Engine_Size', 'Cylinders','Fuel_Consumption_in_City(L/100 km)', 'Fuel_Consumption_in_City_Hwy(L/100 km)'])
DfTr
```

```python
from sklearn import preprocessing

X_normalized = preprocessing.normalize(X_train, norm='l2')
frame = pd.DataFrame(X_normalized, columns = ['Engine_Size','Cylinders','Fuel_Consumption_in_City(L/100 km)','Fuel_Consumption_in_City_Hwy(L/100 km)'])
frame
```

```python
df['Категория'] = pd.cut(df['Smog_Level'], bins=[0, 3, 5, 7], labels=["Низкий", "Средний", "Высокий"])
df
```

```python
df.dtypes
```

```python
df["Категория"].value_counts().sort_index()
```

```python
from sklearn.compose import make_column_selector as selector

categorical_columns_selector = selector(dtype_include='category')
categorical_columns = categorical_columns_selector(df)
categorical_columns
```

```python
df_cat = df[categorical_columns] #отберем только те столбцы в которых присутствует категориальная перемнная
df_cat.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

cat_column = df_cat[["Категория"]]

print(df_cat['Категория'].unique()) #посмотрим какие виды образования присутствуют в выборке

encoder = OrdinalEncoder()   # вызываем кодировщик
encoder  = encoder.fit(df_cat[['Категория']])
df_cat['Категория'] = encoder.transform(df_cat[['Категория']])
df_cat
```

```python
print(encoder.categories_)
print(len(encoder.categories_[0]))
print(df_cat['Категория'].unique())
df_cat['Категория'].unique().size
```
